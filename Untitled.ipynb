```python
import pandas as pd
```

```python
df = pd.read_csv('more_restaurants_in_boston.csv')
df.head()
```

```python
pd.set_option('display.max_columns', None)
sorted_df = df[["name", "rating", 'price_level', 'rating_count', 'tags', 'address']].drop_duplicates(subset=['name', 'rating', 'price_level', 'rating_count', 'address'], keep='first').sort_values(by=['rating'], ascending=False).where(lambda d: d['rating_count'] > 25).dropna(how='all')
sorted_df
```
